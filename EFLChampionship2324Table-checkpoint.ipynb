```python
import pandas as pd
```

```python
df = pd.read_html('https://fbref.com/en/comps/10/Championship-Stats', 
                  attrs={"id": "results2023-2024101_overall"})[0]
```

```python
df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
Squad = df['Squad'].tolist()  # Assuming 'Squad' is the column name for team names
GD = df['GD'].tolist()  # Assuming 'GD' is the column name for goal differences

# Create data dictionary
data = {
    'Squad': Squad,
    'GD': GD 
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df_sorted = df.sort_values(by='GD', ascending=True)
plt.figure(figsize=(8, 6))
plt.scatter(df_sorted['GD'], df_sorted['Squad'], color='red', alpha=0.7)

plt.scatter(df_sorted[df_sorted['GD'] < 0]['GD'], df_sorted[df_sorted['GD'] < 0]['Squad'], color='red', alpha=0.7, label='Negative GD')
plt.scatter(df_sorted[df_sorted['GD'] >= 0]['GD'], df_sorted[df_sorted['GD'] >= 0]['Squad'], color='blue', alpha=0.7, label='Positive GD')
plt.xlim(-80, 80)

plt.xlabel('Goal Differences')
plt.ylabel('Teams')
plt.title('Goal Difference of Teams in 2023/2024 EFL Championship Regular Season')

for i, (gd, squad) in enumerate(zip(df_sorted['GD'], df_sorted['Squad'])):
    plt.text(gd + 2, squad, str(gd), ha='left', va='center')
    
plt.show()
```
